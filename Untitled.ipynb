```python
import sklearn
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('survey_results_public.csv', encoding='1252')
```

```python
df.head()
```

```python
df = df[['Country', 'EdLevel', 'YearsCodePro', 'Employment', 'ConvertedComp']]
df = df.rename({'ConvertedComp':'Salary'}, axis=1)
df.head()
```

```python
df = df[df['Salary'].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df = df[df['Employment'] == 'Employed full-time']
df = df.drop('Employment', axis=1)
df.info()
```

```python
df['Country'].value_counts()
```

```python
#remove countries with only one datapoint since model cannot learn from them
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
#from above, none of the boxes reaches over 0.25 ($250000) hence we define it as the cutoff value

df = df[df['Salary'] <= 250000]
df = df[df['Salary'] >= 10000]
df = df[df['Country'] != 'Other']
```

```python
#to have a better visualisation of data around median

fig, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country',ax=ax)
plt.suptitle('Salary (US$) V Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
#inspect YearsCodePro

df["YearsCodePro"].unique()
```

```python
#change 'less than 1 year' to 0.5 and 'more than 50 years' to 50

def clean_experience(x):
    if x ==  'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)

df["YearsCodePro"].unique()
```

```python
#inspect EdLevel

df["EdLevel"].unique()
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(clean_education)

df["EdLevel"].unique()
```

```python
# convert string features to integer values

from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df["EdLevel"].unique()
```

```python
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df["Country"].unique()
```

```python
#start training
```

```python
#may consider splitting into training and test data*

X = df.drop("Salary", axis=1) #features
y = df["Salary"] #class labels
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X, y.values)
```

```python
y_pred = linear_reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y, y_pred))
```

```python
print(f' linear regression: {error}')
```

```python
# from above, the error is relatively significant hence consider another model such as decisiontree
```

```python
from sklearn.tree import DecisionTreeRegressor

decisiontree = DecisionTreeRegressor(random_state = 0)
decisiontree.fit(X, y.values)

y_pred = decisiontree.predict(X)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
```

```python
print(f' decision tree error: {error}')
```

```python
#consider randomforest
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X, y.values)

y_pred = random_forest_reg.predict(X)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print(f' random forest error : {error}')
```

```python
# determing best parameters using gridsearch

from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X, y.values)

```

```python
regressor = gs.best_estimator_

regressor.fit(X, y.values)
y_pred = regressor.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))

#error higher than the one generated from random forest, may be due to randomness in the tree
```
